```python
# for each data split the word based on space and store it inside a list
def extract_list(inp):
    output=[]
    for i in inp:
        output = inp.split()
    return output    
```

```python
# apply the function for each row in Author col
books['Author'] = books['Author'].apply(extract_list)
```

```python
# apply the function for each row in Publisher col
books['Publisher'] = books['Publisher'].apply(extract_list)
```

```python
books.head(5)
```

```python
# join the list items in a single line
books['Author'] = books['Author'].apply(lambda x:"".join(x))
```

```python
books['Publisher'] = books['Publisher'].apply(lambda x:"".join(x))
```

```python
# join Blurb, Author & Publisher cols in a single col named tags
books['tags'] = books['Blurb'] + " " + books['Author'] + " " + books['Publisher']
```

```python
# take only title and tags from books dataset
new_books = books[['Title','tags']]
```

```python
new_books.sample(4)
```

```python
# take 5000 sample data from new_books dataset
new_books_sample = new_books.sample(5000)
```

```python
new_books_sample.reset_index(inplace = True)
```

```python
new_books_sample.shape
```

```python
new_books_sample.sample(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv  = CountVectorizer(max_features= 5000, stop_words= 'english')
```

```python
cv.fit_transform(new_books_sample['tags'])
```

```python
vector = cv.fit_transform(new_books_sample['tags'])
```

```python
vector.toarray()
```

```python
vectors = vector.toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
#find the index of the seached book
def findIndex(book):
    index_pos = new_books_sample[new_books_sample['Title'] == book].index[0]
    print(index_pos)
```

```python
sorted(list(enumerate(similarity[3])), key = lambda x:x[1], reverse = True)[1:6]
```

```python
def recommend(book):
    index_pos = new_books_sample[new_books_sample['Title'] == book].index[0]
    book_list = sorted(list(enumerate(similarity[index_pos])), key = lambda x:x[1], reverse = True)[1:6]
        
    for item in book_list:
        print(new_books_sample.iloc[item[0]]['Title'])
```

```python
new_books_sample.sample(5)
```

```python
recommend('A Blossom Promise')
```

```python
recommend('The Saggy Baggy Elephant')
```
